## Simulation fundamentals, $C_1$

Simulate $N$ sets of springs

```python
# Fix random generator for repeatability
np.random.rand(42)
# Number of simulations
N = int(1e6)
c1 = 100*(1+0.2*(np.random.rand(N,1)-0.5))
c2 = 150*(1+0.2*(np.random.rand(N,1)-0.5))
c3 = 200*(1+0.2*(np.random.rand(N,1)-0.5))
```

Assemble spring packages

```python
c = 8*c1+8*c2+8*c3
```

Analyse distribution

```python
sns.distplot(c)
```
